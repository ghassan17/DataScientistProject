# Q1: What is the top languages used in Spain over the past 5 years?

```python
def fill_na_with_mode_for_all_columns(df):
    """
    Fill null values in all columns with the mode.

    Parameters:
    - df: pandas DataFrame

    Returns:
    - df: pandas DataFrame, with null values in all columns filled with the mode
    """
    for column_name in df.columns:
        mode_value = df[column_name].mode().iloc[0]  # Calculate the mode of the column
        df[column_name] = df[column_name].fillna(mode_value)  # Fill null values with the mode
    return df

# Example usage:
# Assuming 'your_dataframe' is the DataFrame with null values

cleaned2019 = fill_na_with_mode_for_all_columns(df2019)
cleaned2020 = fill_na_with_mode_for_all_columns(df2020)
cleaned2021 = fill_na_with_mode_for_all_columns(df2021)
cleaned2022 = fill_na_with_mode_for_all_columns(df2022)
cleaned2023 = fill_na_with_mode_for_all_columns(df2023)

cleaned2023
```

```python
Spain2019 = cleaned2019[cleaned2019["Country"]=="Spain"]
Spain2020 = cleaned2020[cleaned2020["Country"]=="Spain"]
Spain2021 = cleaned2021[cleaned2021["Country"]=="Spain"]
Spain2022 = cleaned2022[cleaned2022["Country"]=="Spain"]
Spain2023 = cleaned2023[cleaned2023["Country"]=="Spain"]

Spain2023
```

```python
def count_languages(data, column_name, top_n=10):
    """
    Count the occurrences of programming languages in a specified column of a DataFrame.

    Parameters:
    - data: pandas DataFrame
    - column_name: str, the name of the column containing semicolon-separated languages
    - top_n: int, the number of top languages to display (default is 10)

    Returns:
    - counts_df: pandas DataFrame, sorted by counts in descending order
    """
    language_counts = {}

    # Iterate through each row and count the occurrences of each language
    for languages in data[column_name].dropna():
        # Split the string of languages into a list
        language_list = languages.split(';')

        # Iterate through each language in the list and update the counts
        for language in language_list:
            language = language.strip()  # Remove leading and trailing whitespaces
            if language in language_counts:
                language_counts[language] += 1
            else:
                language_counts[language] = 1

    # Convert the dictionary to a pandas DataFrame for easier manipulation
    counts_df = pd.DataFrame(list(language_counts.items()), columns=['Language', 'Count'])

    # Sort the DataFrame by counts in descending order
    counts_df = counts_df.sort_values(by='Count', ascending=False, ignore_index=True)

    # Return the top_n languages
    return counts_df.head(top_n)

# Example usage:
result_df2023 = count_languages(Spain2023, 'LanguageHaveWorkedWith', top_n=10)
result_df2022 = count_languages(Spain2022, 'LanguageHaveWorkedWith', top_n=10)
result_df2021 = count_languages(Spain2021, 'LanguageHaveWorkedWith', top_n=10)
result_df2020 = count_languages(Spain2020, 'LanguageWorkedWith', top_n=10)
result_df2019 = count_languages(Spain2019, 'LanguageWorkedWith', top_n=10)

print(result_df2023)
```

```python
import matplotlib.pyplot as plt

def plot_language_counts(counts_df, top_n=10):
    """
    Plot a bar chart for the top N programming languages and their counts.

    Parameters:
    - counts_df: pandas DataFrame with 'Language' and 'Count' columns
    - top_n: int, the number of top languages to plot (default is 10)
    """
    # Select the top N languages
    top_languages = counts_df.head(top_n)

    # Plotting the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(top_languages['Language'], top_languages['Count'], color='red')
    plt.xlabel('Programming Language')
    plt.ylabel('Count')
    plt.title(f'Top {top_n} Programming Languages')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
    
    return(top_languages)

# Example usage:
# Assuming 'result_df' is the DataFrame with language counts
print(plot_language_counts(result_df2019, top_n=10))
print(plot_language_counts(result_df2020, top_n=10))
print(plot_language_counts(result_df2021, top_n=10))
print(plot_language_counts(result_df2022, top_n=10))
print(plot_language_counts(result_df2023, top_n=10))

```

# Q2: What is the change of education level over the past 5 years?

```python

def plot_column_counts(data, column_name, top_n=None):
    """
    Plot a bar chart for the count of values in a specified column.

    Parameters:
    - data: pandas DataFrame
    - column_name: str, the name of the column to plot
    - top_n: int or None, the number of top values to plot (default is None)
      If top_n is None, plot all unique values in the column.
    """
    # Count the occurrences of each value in the specified column
    value_counts = data[column_name].value_counts()

    # Select the top N values if specified
    if top_n is not None:
        value_counts = value_counts.head(top_n)

    # Plotting the bar chart
    plt.figure(figsize=(10, 10))
    value_counts.plot(kind='bar', color='skyblue')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title(f'Count of Values in {column_name}')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()

    return value_counts
# Example usage:
# Assuming 'your_dataframe' is the DataFrame and 'your_column' is the column to plot


print(plot_column_counts(cleaned2019, 'EdLevel', top_n=10))
print(plot_column_counts(cleaned2020, 'EdLevel', top_n=10))
print(plot_column_counts(cleaned2021, 'EdLevel', top_n=10))
print(plot_column_counts(cleaned2022, 'EdLevel', top_n=10))
print(plot_column_counts(cleaned2023, 'EdLevel', top_n=10))
```

# Q3: What is the change of learning ways over the past 3 years?

```python

def plot_column_percentages(data, column_name, top_n=None):
    """
    Plot a pie chart for the percentage distribution of values in a specified column.

    Parameters:
    - data: pandas DataFrame
    - column_name: str, the name of the column to plot
    - top_n: int or None, the number of top values to plot (default is None)
      If top_n is None, plot all unique values in the column.
    """
    # Count the occurrences of each value in the specified column
    value_counts = data[column_name].value_counts()

    # Select the top N values if specified
    if top_n is not None:
        value_counts = value_counts.head(top_n)

    # Plotting the pie chart
    plt.figure(figsize=(8, 8))
    value_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title(f'Percentage Distribution of Values in {column_name}')
    plt.ylabel('')  # Remove the default 'count' label
    plt.show()

    return value_counts


print(plot_column_percentages(cleaned2021, 'LearnCode', top_n=3))
print(plot_column_percentages(cleaned2022, 'LearnCode', top_n=3))
print(plot_column_percentages(cleaned2023, 'LearnCode', top_n=3))
```
